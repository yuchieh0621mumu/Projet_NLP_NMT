```python
from bertalign import Bertalign
```

```python
def read_txt_files(src_path, tgt_path):
  with open(src_path, 'r') as src_file:
    src_lines = src_file.readlines()

  with open(tgt_path, 'r') as tgt_file:
    tgt_lines = tgt_file.readlines()

  src_text = "".join(src_lines)
  tgt_text = "".join(tgt_lines)

  return src_text, tgt_text


src_file_path = '../data/ACL.6060.dev.en-xx.en.source.txt'
tgt_file_path = '../data/ACL.6060.dev.en-xx.fr.cible.txt'

src_text, tgt_text = read_txt_files(src_file_path, tgt_file_path)

aligner = Bertalign(src_text, tgt_text)
aligner.align_sents()
```
